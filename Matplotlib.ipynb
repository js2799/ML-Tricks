```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,100,10)
y=x*x
plt.plot(x,y)
```

```python
x=['A','B','C']
y=[50,70,90]
plt.xlabel('Categories')
plt.ylabel('Numbers')
plt.title("Line Chart Example")
plt.plot(x,y)
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
print("X:",x)
print("y:",y)
plt.plot(x,y)
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
y1=np.cos(x)
# print("X:",x)
# print("y:",y)
plt.plot(x,y)
plt.plot(x,y1)
```

```python
#Bar Chart
x=np.array(["A","B","C","D"])
y=np.array([30,40,50,80])
plt.xlabel('Categories')
plt.ylabel("Values")
plt.title("BarChart")
plt.bar(x,y)
```

```python
#horizontal bar graph
x1=np.array(["A","B","C","D"])
y1=np.array([20,50,70,90])
plt.xlabel('Categories')
plt.ylabel("Values")
plt.title("BarChart")
plt.barh(x1,y1)
```

```python
#BarChart with Different Dataset

X=np.array(["A","B","C","D"])
V1=np.array([60,70,80,90])
V2=np.array([30,35,40,45])
V3=np.array([15,17.5,20,22.5])
bar_width=0.35
plt.bar(X, V1,bar_width, label="DataSet1",color='orange')
plt.bar(X, V2,bar_width, label="DataSet2",color='white')
plt.bar(X, V3,bar_width, label="DataSet3",color='green')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Group Bar Chart")
plt.legend()
```

```python
cat=np.array(['A','B','C','D'])
mat=np.array([50,20,10,5])
explode=(0,0.2,0,0.2)
plt.pie(mat, labels=cat, explode=explode, autopct="%1.1f5%%")
plt.plot(cat,mat)
```

```python
shop=np.array(['Electronics','Cloths','Grocery','Household'])
value=np.array([5000,6000,4000,3500])
explode=(0,0.1,0,0)
plt.pie(value, labels=shop, explode=explode,  autopct="%1.1f%%", shadow=True)
# plt.legend()
```
