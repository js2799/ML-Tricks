# Matplotlib


```python
import matplotlib.pyplot as plt
```

```python
ma =pd.read_csv('Movie-Data.csv')
```

```python
ma.columns
```

```python
ma.nunique()
```

```python
#Line Chart

x=ma['Title']
y=ma['Runtime (Minutes)']

plt.title('LineChart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot(x,y)
```

```python
#Bar Chart

xx=np.array(ma['Year'])
yy=np.array(ma['Votes'])
plt.title("Bar Chart")
plt.xlabel('Year')
plt.ylabel('Votes')
plt.bar(xx,yy)
# plt.barh(xx,yy) #Horinzontal Bar Chart
```

```python
##Group BarChart

X=np.array(ma['Revenue (Millions)'])
V1=np.array(ma['Year'])
V2=np.array(ma['Votes'])
V3=np.array(ma['Rating'])
bar_width=0.35
plt.bar(X,V1,bar_width,label="DataSet1",color='blue')
plt.bar(X,V2,bar_width,label="DataSet2",color='red')
plt.bar(X,V3,bar_width,label="DataSet3",color='green')
plt.title("Group Bar Chart")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
```

```python
#piechart

xxx=np.array(ma['Year'])
yyy=np.array(ma['Rating'])
plt.pie(yyy,labels=xxx,autopct='%1.1f5%%')
```
